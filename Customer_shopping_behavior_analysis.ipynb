```python
# now  we will fill NAN values with median of each category ...
# here for clothing we will fill median of clothing in review rating .if we fill median with total median it results false...
#df["Review Rating"]
#df.groupby("Category")["Review Rating"]
#df["Review Rating"].value_counts()
df["Category"].value_counts()
```

```python
# medain of each category
df.groupby("Category")["Review Rating"].median()

```

```python
#Instead of filling all missing values with:
#overall mean ❌
#overall median ❌
#We fill based on category-wise logic ✅
#Which is more accurate.*/

    
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
# nulls are replaced with median values of their category of their median()..
df.isnull().sum()
```

```python
# formating the column names  
# removin spaces and replacing with underscores..
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_")
```
